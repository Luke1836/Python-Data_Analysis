```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
headers = ["fLength", "fWidth", "fSize", "fConc", "fConc1", "fAsym", "fM3Long", "fM3Trans", "fAlpha", "fDist", "class"]
data_train = pd.read_csv("../Sources/magic04.data", names=headers)
```

```python
# data_train["class"].unique() : we get the unique values as an array
data_train["class"] = data_train["class"].str.strip().str.lower()
data_train["class"] = (data_train["class"]=="g").astype(int)
data_train
# If it is gamma signal then it is 1, otherwise (hadron) it is 0.
```

# So now we are going to predict whether the given input or data belongs to class "g" or "h".
- We are going to do classification which is a part of the supervised learning.
- Here we are given a properly labelled data.
- We make sure to give the info we entered a proper classification based on the class.

```python
for labels in headers[:-1]:
    plt.hist(data_train[data_train['class'] == 1][labels], color='blue', label='gamma', alpha=0.7, density=True)
    plt.hist(data_train[data_train['class'] == 0][labels], color='red', label='hadron', alpha=0.7, density=True)
    plt.legend()
    plt.title(labels)
    plt.xlabel(labels)
    plt.ylabel('probability')
    plt.show()
```
